```python
from sklearn.cluster import KMeans#### import the  package
```

```python
cluster = KMeans(n_clusters=3)###  we have fit our model in cluster
cluster.fit(x)

p= cluster.predict(x)### predict the cluster
p , len(p)
```

```python
wcss = []
for i in range(1,10):
    km = KMeans(n_clusters=i, max_iter=200)
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
wcss

plt.plot(range(1, 10), wcss)## visualize the plot
plt.show()
```

### fiting the KMeans to data set

```python
cluster = KMeans(n_clusters=5)
cluster.fit(x)

p1= cluster.predict(x)
p1 , len(p1)
```
